```python
import numpy as np
import pandas as pd
```

```python
mobile = pd.read_csv(r"mobile.csv")
```

```python
# 1
mobile['First Name'] = mobile['Customer Name'].apply(lambda value: value.split()[0])
print(mobile)
```

```python
# 2
mobile['Cell Phone Company'] = mobile['Phone Number'].apply(lambda value: 'Pelephone' if value.split('-')[0] == '050' else ('Golan Telecom' if value.split('-')[0] == '051' else ('Cellcom' if value.split('-')[0] == '052' else ('Cellcom' if value.split('-')[0] == '053' 'Motorola' else 'Partner'))))
print(mobile)
```

```python
# 3.a Cell Phone Company Customer Name
print('3.a')
print(mobile['Cell Phone Company'].value_counts())

# b
print('\n3.b')
print(mobile['Cell Phone Company'].nunique())

# c
print('\n3.c')
print(mobile['Cell Phone Company'].unique())
```

```python
# 4.a 
print('4.a')
mobile['Customer type'] = mobile.apply(lambda row: 'talker' if row['Talked Minutes'] > row['Internet Minutes'] else 'surfer', axis=1)
print(mobile)

# b
print('\n4.b')
mobile['Customer type v'] = np.vectorize(lambda talked, internet: 'talker' if talked > internet else 'surfer')(mobile['Talked Minutes'], mobile['Internet Minutes'])
print(mobile)
```

```python
# 5
print(mobile.sort_values(['Talked Minutes'], ascending=True))
```

```python
# 6
print(mobile.sort_values(['Talked Minutes', 'Internet Minutes'], ascending=[True, False]))
```

```python
# 7
mobile.nlargest(5, 'Talked Minutes')
```

```python
# 8
mobile.nsmallest(3, 'Internet Minutes')
```

```python
# 9.a
print('9.a')
duplicate_rows = mobile[mobile.duplicated()]
print(duplicate_rows)

# b
print('\n9.a')
print(mobile.drop_duplicates())

```
